```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as img
from PIL import Image
import numpy as np
from PIL import ImageFilter
```

```python
print('if i =0 Image read and display\n')
print('if i =1 Pseudo color Image\n')
print('if i =2 Pseudo color Image color bar\n')
print('if i =3 Image Resizing\n')
print('if i =4 Image Interpolation\n')
print('if i =5 RGB to Gray Image\n')
print('if i =6 Histogram Plot\n')
print('if i =7 Cropping a Portion of an Image\n')
print('if i =8 Shape of an Image and gray scale conversion\n')
print('if i =9 Image transform\n')
print('if i =10 Image Filtering\n')
print('if i=11 Image Details and Changing Image File Format')
i = eval(input('Enter the Digital Image Processing program you want to run=\n'))
```

```python

if (i==0):
    print('Image and Display')
    im = img.imread('/content/Bird5.jpg')
    plt.imshow(im)
    plt.show()

```

```python
if  (i==1):
    print('Pseudo Color Image')
    im = img.imread('/content/Bird5.jpg')
    Pseudo_Image = im[:,:,0]
    plt.imshow(Pseudo_Image)
    plt.title('Pseudo Color Image')
    plt.show()

```

```python
if (i==2):
    print('Pseudo color image with color bar')
    im = img.imread('/content/Bird5.jpg')
    Pseudo_Image = im[:,:,0]
    plt.imshow(Pseudo_Image,cmap='hot')
    plt.title('Pseudo Color Image with colorbar')
    plt.colorbar()
    plt.show()

```

```python
if (i==3):
    print('Image Resizing')
    im = Image.open('/content/Bird5.jpg')
    im.thumbnail((128,128),Image.ANTIALIAS)
    plt.imshow(im)
    plt.title('Image 128x128 resized')
    plt.show()

```

```python
if (i==4):
    print('Image Interpolation')
    im = Image.open('/content/Bird5.jpg')
    im.thumbnail((128,128),Image.ANTIALIAS)
    fig, axs = plt.subplots(1,2)
    axs[0].imshow(im)
    axs[0].set_title('Image 128x128 without interpolation')
    axs[1].imshow(im)
    axs[1].imshow(im,interpolation='sinc')
    axs[1].set_title('Image 128x128 with interpolation')
    plt.show()

```

```python
if (i==5):
    print('Grayscale Image')
    im = Image.open('/content/Bird5.jpg')
    im1 = Image.open('/content/Bird5.jpg').convert('LA')
    im1.save('im_grayscale.png')
    im2 = img.imread('/content/im_grayscale.png')
    fig, axs = plt.subplots(1,2)
    axs[0].imshow(im)
    axs[0].set_title('Colour Image')
    axs[1].imshow(im2)
    axs[1].set_title('GrayScale Image')
    plt.show()


```

```python
if (i==6):
    print('Histogram Plot')
    im = Image.open('/content/Bird5.jpg').convert('LA')
    im.save('im_grayscale.png')
    im1 =  img.imread('/content/im_grayscale.png')
    plt.figure(1)
    plt.imshow(im1)
    plt.title('grayscale image')
    plt.figure(2)
    plt.hist(im1.ravel(),bins=256)
    plt.title('Histogram Plot')
    plt.show()

```

```python
if (i==7):
    print('Cropping a portion from an Image')
    im = Image.open('/content/Bird5.jpg')
    box =(50,50,200,200)
    Croped_Image = im.crop(box)
    plt.imshow(Croped_Image)
    plt.title('CroppedImage from Original Image')
    plt.show()

```

```python
if (i==8):
    print('1.To print shape of an Image')
    print('2.To display RED,GREEN,BLUE images separately')
    print('3.To reconstruct original image from R,G,B matrices')
    im =img.imread('/content/Senthil.png')
    im2 = Image.open('/content/Senthil.png').convert('LA')
    im2.save('Senthil_grayscale.png')
    im1 = img.imread('/content/Senthil_grayscale.png')
    plt.figure(1)
    plt.imshow(im)
    print('Colour Image Shape')
    #print(im.shape)
    plt.figure(2)
    plt.imshow(im1)
    print('Grayscale Image Shape')
   # print(im1.shape)
    m = len(im)
    n = len(im[0])
    s =(m,n)
    im2= np.zeros(s)
    plt.figure(3)
    for i in range(len(im)):
        for j in range(len(im[0])):
            im2[i][j] = im[i][j][0]
    
    plt.imshow(im2)
    plt.title('Red Matrix')
    im3= np.zeros(s)
    plt.figure(4)
    for i in range(len(im)):
        for j in range(len(im[0])):
            im3[i][j] = im[i][j][1]
    
    plt.imshow(im3)
    plt.title('Green Matrix')
    im4= np.zeros(s)
    plt.figure(5)
    for i in range(len(im)):
        for j in range(len(im[0])):
            im4[i][j] = im[i][j][2]
    
    plt.imshow(im4)
    plt.title('Blue Matrix')
    s1=(m,n,3)
    im5 = np.zeros(s1)
    plt.figure(6)
    for k in range(3):
        for i in range(len(im)):
            for j in range(len(im[0])):
                if (k==0):
                    im5[i][j][k] = im2[i][j]
                elif (k==1):
                    im5[i][j][k] = im3[i][j]
                elif (k==2):
                    im5[i][j][k] = im4[i][j]
                    
    
    plt.imshow(im5)
    plt.title('Reconstructed Colour Images from RED GREEN BLUE  Matrices')
    #print(im.shape)
    plt.show()
```

```python
if (i==9):
    print('Image Transform')
    im = Image.open('/content/Bird5.jpg')
    out1 = im.transpose(Image.FLIP_LEFT_RIGHT)
    out2 = im.transpose(Image.FLIP_TOP_BOTTOM)
    out3 = im.transpose(Image.ROTATE_90)
    out4 = im.transpose(Image.ROTATE_180)
    out5 = im.transpose(Image.ROTATE_270)
    plt.figure(1)
    plt.imshow(out1)
    plt.title('Flip LEFT-RIGHT')
    plt.figure(2)
    plt.imshow(out2)
    plt.title('Flip TOP-BOTTOM')
    plt.figure(3)
    plt.imshow(out3)
    plt.title('Rotate 90')
    plt.figure(4)
    plt.imshow(out4)
    plt.title('Rotate 180')
    plt.figure(5)
    plt.imshow(out5)
    plt.title('Rotate 270')
    plt.show()
```

```python
if (i==10):
    print('Image Filtering')
    im = Image.open('/content/Bird5.jpg')
    im1 = im.filter(ImageFilter.BLUR)
    im2 = im.filter(ImageFilter.MinFilter)
    im3 = im.filter(ImageFilter.MedianFilter)
    im4 = im.filter(ImageFilter.UnsharpMask)
    plt.figure(1)
    plt.imshow(im1)
    plt.title('Gaussian Blur Filter')
    plt.figure(2)
    plt.imshow(im2)
    plt.title('Minimum Filter')
    plt.figure(3)
    plt.imshow(im3)
    plt.title('Median Filter')
    plt.figure(4)
    plt.imshow(im4)
    plt.title('Unsharp Mask Filter')
    plt.show()
```

```python
if (i==11):
    print('Image Details and Changing Image File Format')
    im = Image.open('/content/Bird5.jpg')
    print('image  format',im.format)
    print('image mode',im.mode)
    print('image size',im.size)
    print('image palette',im.palette)
    print('Changing Image file format from png to jpg')
    im1 = Image.open('Full_Size.jpg')
    im1.save('Full_Size.png')
    im2 = Image.open('Full_Size.png')
    plt.figure(1)
    plt.imshow(im1)
    plt.title('Full_Size.jpg')
    plt.figure(2)
    plt.imshow(im1)
    plt.title('Bird5.jpg')
    plt.imshow(im2)
    plt.title('Full_Size.png')
    plt.show()
```
