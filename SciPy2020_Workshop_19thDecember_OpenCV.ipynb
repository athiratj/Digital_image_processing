```
# Splitting image gives blue green red channels
B,G,R = cv2.split(img)
```

```
# Merging the individual channels into image
img2 = cv2.merge([B,G,R]) 
```

```
cv2_imshow(img2)
print('Merged Image')
```

```
grey = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
cv2_imshow(grey)
print('Gray Image')
```

```
grey_3_channel = cv2.cvtColor(grey, cv2.COLOR_GRAY2BGR)
numpy_vertical = np.vstack((img,grey_3_channel))
numpy_horizontal = np.hstack((img,grey_3_channel))
cv2_imshow(img)
print("original image")
cv2_imshow(grey)
print("gray image")
cv2_imshow(numpy_vertical)
print("Numpy Vertical")
cv2_imshow(numpy_horizontal)
print("Numpy Horizontal")
```

```
img3 = cv2.imread('/content/Bird2.jpg')
height,width =img3.shape[0:2]
rotationMatrix = cv2.getRotationMatrix2D((width/2, height/2),270,1)
rotatedImage = cv2.warpAffine(img3,rotationMatrix,(width, height))
cv2_imshow(img3)
print('originalImage')
cv2_imshow(rotatedImage)
print('rotated to 270 degrees')
```

```
height,width = img3.shape[0:2]
startRow =int(height*0.2)
endRow = int(height*0.8)
startCol = int(width*0.2)
endCol = int(width*0.8)
croppedImage = img3[startRow:endRow, startCol:endCol]
cv2_imshow(img3)
print('Original Image')
cv2_imshow(croppedImage)
print('Cropped Image')
```

```
font= cv2.FONT_ITALIC
img4=cv2.putText(img,'1',(10,50),font,2,(255,255,266))
cv2_imshow(img4)
```
